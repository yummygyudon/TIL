```python

from selenium import webdriver

driver = webdriver.Chrome('C:/Temp/chromedriver')
driver.implicitly_wait(3) 

driver.get("https://www.megabox.co.kr/movie-detail/comment?rpstMovieNo=21033000")
import time

scoreList = []
themeList = []
commentList = []
time.sleep(3)
stopFlag = False

while True : 
    for k in range(1,31):
        time.sleep(3)
        
        for i in range(2,12) :
            score = driver.find_element_by_css_selector('#contentData div.movie-idv-story li:nth-child(%d) div.story-cont > div.story-point > span'%i)
            scoreList.append(score.text)
            theme = driver.find_element_by_css_selector('#contentData div.movie-idv-story li:nth-child('+str(i)+') div.story-cont > div.story-recommend')
            themeList.append(theme.text)
            comment = driver.find_element_by_css_selector('#contentData div.movie-idv-story li:nth-child(%d) div.story-cont > div.story-txt'%i)
            commentList.append('"'+comment.text+'"')
            
        time.sleep(3)

        
        if k == 30 :
            stopFlag = True
            break
        elif (k%10) == 0 :
            NextLink = driver.find_element_by_css_selector('#contentData > div > div.movie-idv-story > nav > a.control.next')
            NextLink.click()
        else :
            NextLink =driver.find_element_by_link_text(str(k+1))
            NextLink.click()
    if stopFlag == True :
        break


print(len(scoreList), len(themeList), len(commentList))
        
with open('../output/movie.csv', "wt", encoding="utf-8") as f:
    f.write('score,genre,review\n')
    for i in range(len(scoreList)):
        f.write(scoreList[i]+","+themeList[i]+","+commentList[i]+'\n')  
    f.close()
    
    
#### 페이지 전체 받아와서 zip() 함수 사용하는 것도 가능
```
