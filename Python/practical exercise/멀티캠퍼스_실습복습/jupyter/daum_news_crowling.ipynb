```python
import requests
from bs4 import BeautifulSoup
import re

title = []
comname = []
url = 'https://news.daum.net/ranking/popular/'
r = requests.get(url)
bs = BeautifulSoup(r.text, 'html.parser')

titleList = bs.select('#mArticle > div.rank_news > ul.list_news2 div.cont_thumb > strong > a')
comnameList = bs.select('#mArticle > div.rank_news > ul.list_news2 div.cont_thumb > strong > span')

for titleDom in titleList:
    article = titleDom.text
    article = re.sub('\'', '', article)
    title.append(article)
for comnameDom in comnameList:
    comname.append(comnameDom.string)
    
print("-- 기사 제목 --")
print(title, len(title))
print("-- 기사 출처 뉴스사 --")
print(comname, len(comname))

with open('output/news.csv', "wt", encoding="utf-8") as f:
    f.write('newstitle,newscomname\n')
    for i in range(len(title)):
        f.write(title[i]+","+comname[i]+'\n')  
```
