```python
import pandas as pd
import numpy as np

# 1
data = {
    'name':['둘리', '또치', '도우너', '희동이'],
    'kor':[90, 80, 70, 70],
    'eng':[99, 98, 97, 46],
    'mat':[90, 70, 70, 60],
}
df = pd.DataFrame(data)

df["class"] = ['1반', '2반', '1반', '2반']
print(df)


# 2
new_stu = ['마이콜',80, 80, 80, '1반']
df.loc[4] = new_stu

print(df)


#3
df = df.set_index("name")
df.rename(columns={'kor':'국어','eng':'영어','mat':'수학','class':'반번호'}, inplace=True)
#df.columns=['국어', '영어', '수학', '반번호']

print(df)


# 4
df.loc['마이콜','국어':'수학'] = [100, 100, 100]
df.loc['희동이','영어'] = 90

print(df)


# 5
df = df.reset_index()
print(df)

df.rename(columns={'name':'성명'}, inplace=True)
print(df)


# 6
df1 = df.sort_values(by="국어", ascending=False)
print(df1)

df2 = df.sort_values(by="영어")
print(df2)


# 7
total = df['국어']+df['영어']+df['수학']

df['총점'] = total

df = df.sort_values(by="총점", ascending=False)
print(df)


# 8
#df3 = df[:] _ df복사는 슬라이싱보다 .copy()가 더 바람직한 방법이라고 함.
df3 = df.copy()
df3.drop(['반번호'],axis=1,inplace=True)
print(df3)
print(df)


# 9
df4 = df.copy()
df4.drop(4,inplace=True)
print(df4)
print(df)
```

```python
# 10
data1 = {'kor':90, 'mat':80}
data2 = {'kor':90, 'eng':70}
data3 = {'kor':90, 'eng':70, 'mat':80}
series1 = pd.Series( data1 )
series2 = pd.Series( data2 )
series3 = pd.Series( data3 )
result_2to1 = series1.add(series2, fill_value=0)
result = result_2to1.add(series3, fill_value=0)
print(result)
```

```python
# 11
data = {
    'X1':[2.9, 2.4, 2, 2.3, 3.2],
    'X2':[9.2, 8.7, 7.2, 8.5, 9.6],
    'X3':[13.2, 11.5, 10.8, 12.3, 12.6],
    'X4':[2, 3, 4, 3, 2]
} 
df = pd.DataFrame(data, index=['Y1','Y2','Y3', 'Y4', 'Y5'])


#첫번째 결과
df.loc['Y6'] = [10,20,30,40]
print(df)

#두번째 결과
df= df+10
print(df)

#세번째 결과
tot = df['X1']+df['X2']+df['X3']+df['X4']
df['total']= tot
print(df)

#네번째 결과
df = df.T
print(df)
```
