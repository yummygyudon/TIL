```python
import pandas as pd

df1 = pd.read_csv("../data/score.csv")

print("컬럼명 : ",df1.columns)
print("인덱스 : ",df1.index)
df1['total'] = df1['kor']+df1['eng']+df1['mat']
df1['avg'] = round(df1['total']/(len(df1.columns)-2),2)
display(df1)
```
