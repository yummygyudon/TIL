```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = "C:/JDG(MultiCampus)/PYDATAexam/data/210 옴니고딕 030.ttf"   
font_name = font_manager.FontProperties(fname=font_path).get_name() 
#print(font_name)
rc('font', family=font_name)

#1
df =pd.DataFrame({
    '1분기' : [500, 690, 1100, 1500, 1990, 1020],
    '2분기' : [450, 700, 1030, 1650, 2020, 1600],
    '3분기' : [520, 820, 1200, 1700, 2300, 2200],
    '4분기' : [610, 900, 1380, 1850, 2420, 2550]
}, index= range(2015,2021,1))
display(df)
```

```python
#2
df.rename(index={2015:'2015년',2016:'2016년', 2017:'2017년', 2018:'2018년', 2019:'2019년', 2020:'2020년'}, inplace=True)
plt.figure(figsize=(10,6))
plt.title("시각화 과제 1", c='m')
plt.plot(df)
plt.legend(df.columns,loc=0)
plt.savefig("../output/hw1.png")
plt.show()
```

```python
#3
plt.figure(figsize=(10,6))
plt.plot(df.T,"^-")
plt.legend(df.index,loc=0)
plt.savefig("../output/hw2.png")
plt.show()
```

```python
#4
plt.figure(figsize=(10,6))
data = [sum(df['1분기']),sum(df['2분기']),sum(df['3분기']),sum(df['4분기'])]
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['red', 'green', '#990000', 'purple']
plt.pie(data, labels=df.columns, autopct='%.1f%%', explode=explode, shadow=True, colors= colors)
plt.savefig("../output/hw3.png")
plt.show()
```

```python
#5
plt.figure(figsize=(10,6))
data = [sum(df.T['2015년']),sum(df.T['2016년']),sum(df.T['2017년']),sum(df.T['2018년']),sum(df.T['2019년']),sum(df.T['2020년'])]

colors = ['r', 'g', '#990000', 'purple', 'yellow', 'pink']
plt.pie(data, labels=df.T.columns, autopct='%.1f%%', colors= colors)
plt.savefig("../output/hw4.png")
plt.show()
```
