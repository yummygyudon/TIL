```python
from selenium import webdriver
import time

driver = webdriver.Chrome('C:/Temp/chromedriver')
driver.implicitly_wait(3) 

driver.get("http://gs25.gsretail.com/gscvs/ko/products/event-goods")
time.sleep(1)

TwoOneBtn = driver.find_element_by_css_selector('#contents > div.cnt > div.cnt_section.mt50 > div > div > ul > li:nth-child(2) > span')
TwoOneBtn.click()

nameList, priceList = [], []
next_page = 1
time.sleep(1)

import re

while True :
    print("%d페이지 시작"%next_page)
    goods = driver.find_elements_by_css_selector('#contents > div.cnt div:nth-child(5) li')
    print(len(goods))
    for i in range(1,len(goods)+1):
        goodsname = driver.find_element_by_css_selector('#contents > div.cnt div:nth-child(5) li:nth-child(%d) p.tit'%i)
        nameList.append(goodsname.text)
        
        goodsprice = driver.find_element_by_css_selector('#contents > div.cnt div:nth-child(5) li:nth-child(%d) p.price > span'%i)
        goodsprice = re.sub("[원|,]",'',goodsprice.text)
        priceList.append(goodsprice)
        
        time.sleep(1)
    
    
    
    if len(goods) < 8 :
        break
    else :
        if (next_page%10 == 0) :
            NextLink =driver.find_element_by_css_selector('#contents > div.cnt div:nth-child(5) > div > a.next')
            next_page += 1
            NextLink.click()
            time.sleep(1)
        else :
            NextLink =driver.find_element_by_link_text(str(next_page+1))
            next_page += 1
            NextLink.click()
            time.sleep(1)


print(len(nameList), len(priceList))
with open('../output/gs25_twotoone.csv', "wt", encoding="utf-8") as f:
   f.write('goodsname,goodsprice\n')
   for i in range(len(nameList)):
       f.write(nameList[i]+","+priceList[i]+'\n')  
   f.close()
    
```
