```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
## 1
iris = sns.load_dataset("iris")
iris_x = iris.loc[:,['sepal_length', 'sepal_width',
            'petal_length', 'petal_width']]
import random
random.seed(1)
for col in range(4):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
iris_x.head(10)
```

```python
#1
display(iris_x.head(10).dropna(subset=iris_x.columns,how='any',axis=0))
```

```python
#2
display(iris_x.dropna(subset=iris_x.columns,thresh=2,axis=0).head(10))
```

```python
#3
display(iris_x.dropna(subset=['sepal_length','sepal_width'],axis=0).head(10))
```

```python
#4
display(iris_x.fillna(0).head(10))
```

```python
#5
iris_copy = iris_x.copy()
#iris_copy['sepal_length'].fillna(0, inplace=True)
#iris_copy['sepal_width'].fillna(0, inplace=True)
#iris_copy['petal_length'].fillna(0, inplace=True)
#iris_copy['petal_width'].fillna(0, inpla=True)
iris_mean = iris_copy.mean(axis=0)
iris_copy.fillna(iris_mean.round(1), inplace=True)
display(iris_copy.head(10))
```

```python
#6
iris_copy1 = iris_x.copy()
iris_copy1.fillna(10, inplace=True)
display(iris_copy1.head(10))
```

```python
#7
iris_copy2 = iris_x.copy()
iris_copy2.fillna(method='ffill', inplace=True)
display(iris_copy2.head(10))
```
